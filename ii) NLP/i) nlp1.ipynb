```python
#How multiple lines defined in python.
# give the no.of ' ' as the same as the no.of lines

sen='''The Natural Language Toolkit (NLTK) is an open source Python library
    for Natural Language Processing.  A free online book is available.'''
#if there is more than one line use 3 single quotes
```

```python
#identification of stop words
from nltk.corpus import stopwords
sp=stopwords.words('english')  #create object for stopword
sp
```

```python
#tokenization
sen='''The Natural Language Toolkit (NLTK) is an open source Python library
    for Natural Language Processing.  A free online book is available.'''
from nltk.tokenize import word_tokenize
words=word_tokenize(sen)
words
```

```python
#We have to convert all the words into lower after tokenisation,then only the stop words remove.
# In above sen A is also a stop word,but it is in uppercase so we've to convert it into lower.
remove_stop=[i.lower() for i in words if i.lower() not in sp]  #to remove all stop words from sen.also convert to lower case.
remove_stop
```

```python
#another method to do tokenization
#ngram
from nltk.util import ngrams
sample='this is a very good book to  study'
data=ngrams(sequence=nltk.word_tokenize(sample),n=2)  #n=2===>2-gram
for i in data:
    print(i)
```

```python
#we always prefer stemming over lemmatization
#stemming
#Porter Stemmer
from nltk.stem import PorterStemmer
ps=PorterStemmer()
words=['programming','reached','walking','running','thought','eated','worked','travelling']
for i in words:
    print(i,":",ps.stem(i))
```

```python
#snowball stemmer
from nltk.stem import SnowballStemmer
sb=SnowballStemmer('english') #for snowball stemmer we have to mention the language
for i in words:
    print(i,":",sb.stem(i))
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet') #we have to download 2 resources here
nltk.download('omw-1.4')
lemmatizer=WordNetLemmatizer()
print('rocks:',lemmatizer.lemmatize('rocks'))
print('books:',lemmatizer.lemmatize('books'))
print('reaches:',lemmatizer.lemmatize('reaches'))
print("walking:",lemmatizer.lemmatize("walking"))
print("running:",lemmatizer.lemmatize("running"))
print("walked:",lemmatizer.lemmatize("walked"))
print("coloures:",lemmatizer.lemmatize("coloures"))

#disadvantage of lemmatization:it only cut the words with 's',wont work for 'ing','ed','es'
```
