```python
#Identification of stop words
nltk.download('stopwords') #it is used for the first time only
from nltk.corpus import stopwords
sp=stopwords.words('english') #Object creation, we have to use which language stop word we want
sp
```

```python
#Tokenization ==> converting into word by word

nltk.download('punkt_tab') #it is used at one time only
sentence = 'Luminar technolab is IT finishing school at kakkanad'
from nltk.tokenize import word_tokenize # we can use TweetTokenizer also instead of TweetTokenizer
words=word_tokenize(sentence)
words 
```
