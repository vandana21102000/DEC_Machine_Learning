```python
df.drop(['Cabin'],axis=1,inplace=True)
df
```

```python
#missing value
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Pclass']=df['Pclass'].fillna(df['Pclass'].mode()[0])
```

```python
#1.	What is the survival rate for males vs. females?
df.groupby('Sex')['Survived'].mean()
```

```python

sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate: Males vs. Females')
plt.ylabel('Survival Rate')
plt.xlabel('Sex')
plt.show()
```

```python
#2.	Did passengers in higher classes (1st class) have a higher survival rate compared to lower classes (3rd class)?

df.groupby('Pclass')['Survived'].mean()

#Passengers in 1st class had the highest chance of survival.

#Passengers in 3rd class had the lowest survival rate.
```

```python
#3.	Did passengers with more family members (siblings/spouses or parents/children) aboard have a higher or lower chance of survival?

df['FamilySize'] = df['SibSp'] + df['Parch']
df.groupby('FamilySize')['Survived'].mean()
```

```python
#4.	Did the port of embarkation (C, Q, S) influence the survival rate?

survival_by_embarked = df.groupby('Embarked')['Survived'].mean()
```

```python


survival_by_embarked.plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title('Survival Rate by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

```python
x=df3.drop(['Survived'],axis=1)
y=df3['Survived']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#normalisathion
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#Model Creation

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
nav=GaussianNB()
sup=SVC()
dtc=DecisionTreeClassifier(criterion='entropy')
rf=RandomForestClassifier(random_state=0,n_estimators=100, class_weight='balanced')
lst=[knn,nav,sup,dtc,rf]

for model in lst:
    print(model)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

params = {'n_neighbors': [3, 5, 7, 9]}
grid = GridSearchCV(KNeighborsClassifier(), params, cv=5)

grid.fit(x_train,y_train)
```

```python
#5)Did passengers who paid higher fares have a better chance of survival?

# Calculate quartiles of Fare
fare_quartiles = df['Fare'].quantile([0.25, 0.5, 0.75, 1.0])
print(fare_quartiles)


fare_bins = [0, 7.91, 14.454, 31, df['Fare'].max()]
fare_labels = ['Low Fare', 'Mid-Low Fare', 'Mid-High Fare', 'High Fare']
df['FareCategory'] = pd.cut(df['Fare'], bins=fare_bins, labels=fare_labels, include_lowest=True)

survival_by_fare_cat = df.groupby('FareCategory')['Survived'].mean().reset_index()

sns.barplot(x='FareCategory', y='Survived', data=survival_by_fare_cat)
plt.title('Survival Rate by Fare Categories')
plt.xlabel('Fare Category')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```
