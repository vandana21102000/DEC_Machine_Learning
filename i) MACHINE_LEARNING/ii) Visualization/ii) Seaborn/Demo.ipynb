```python
#to identify outliers we use boxplot.

#the line in the lower part is the minimum
#the first line in the box is called 25 percentle(Q1)
#the second (center) line in the box is called 50 percentle(Q2)
##the third line in the box is called 75 percentle(Q3)
#the line in the upper part is the maximum
```

```python
#3,7,1,1,8,4,6,9

#1) Sort the datas in ascending order 1,1,3,4,6,7,8,9
#2) Calculate the median it is even so divide it into two 1,1,3,4  6,7,8,9   median=(4+6)/2=5
# this median is the Q2. ie, Q2 = 5
# for Q1, we've to find the median of first part (1,1,3,4) ie, median= (1+3)/2=2 ===> Q1 = 2
# for Q3, we've to find the median of second part (6,7,8,9) ie, median= (7+8)/2=7.5 ===> Q3 = 7.5

#3) Find IQR (Inter Quartile Range) = Q3-Q1
#ie, IQR = Q3 - Q1 = 7.5-2 = 5.5

#4) Calculate top outliers
# Top_outliers = (Q3+(1.5*IQR)) = (7.5+(1.5*5.5)) = 15.75
#here the values above 15.75 is consider as the outliers

#5) Calculate bottom outliers
# Bottom_outliers = (Q1-(1.5*IQR)) = (2-(1.5*5.5)) = -6.25
#here the values below -6.25 is consider as the outliers
```

```python
# Implement the above concept into a code.

import numpy as np
Q1=np.percentile(df['total_bill'],25,method='midpoint')  #Q1
Q3=np.percentile(df['total_bill'],75,method='midpoint')  #Q3
IQR = Q3 - Q1
print(IQR)
```

```python
upper = Q3+1.5*IQR
upper_array = np.array(df['total_bill'] >= upper) 
print('Upper_Bound',upper)
print(upper_array)
print(upper_array.sum())
```

```python
#Below Lower Bound

lower = Q1-1.5*IQR
lower_array = np.array(df['total_bill'] <= lower) 
print('Lower_Bound',lower)
print(lower_array)
print(lower_array.sum())
```

```python
# Removing Oultier funtion( top outlier)
import matplotlib.pyplot as plt
def removal_box_plot(df, column, threshold): #arguments
    sns.boxplot(df[column]) #df['total_bill]  #sns.boxplot(df['toatl_bill])
    plt.title(f'Original Box Plot of {column}')  #40
    plt.show()
    removed_outliers = df[df[column] <= threshold]  #

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers


threshold_value = 40

no_outliers = removal_box_plot(df, 'total_bill', threshold_value)  #()
```
