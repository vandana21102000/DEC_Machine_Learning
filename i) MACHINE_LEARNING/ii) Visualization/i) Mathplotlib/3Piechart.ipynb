```python
# pie chart : working is based on values(%)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data1=['grape','orange','banana','apple']
data=[10,20,30,40]
plt.pie(data,labels=data1,autopct='%.1f%%')
plt.legend(loc='upper left')  #for labeling
plt.tight_layout() # reduce overlap
```
