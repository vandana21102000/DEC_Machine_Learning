```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Gender']=label.fit_transform(df['Gender'])
df['Education_Level']=label.fit_transform(df['Education_Level'])
df['Preferred_Activities']=label.fit_transform(df['Preferred_Activities'])
df['Location']=label.fit_transform(df['Location'])
df['Favorite_Season']=label.fit_transform(df['Favorite_Season'])
```

```python
X=df.drop(columns=['Preference'])
X
```

```python
y=df['Preference']
y
```

```python
# pip install imbalanced_learn 
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
X_resampled,y_resampled=smote.fit_resample(X,y)
X_resampled 
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.naive_bayes import BernoulliNB # used for BernoulliNB because negative values present
b=BernoulliNB()
b.fit(x_train,y_train)
y_pred=b.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
nav=GaussianNB()
sup=SVC()
lst=[knn,nav,sup]
```

```python
for i in lst:
    print("Model name is*************",i)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("perfomance evaluation is ******")
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    print(accuracy_score(y_test,y_pred))
```
