```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
df
```

```python
df0=df['month_number']
```

```python
df1=df['total_profit']
```

```python
# ................... 1 ..................#
plt.plot(df0,df1)
plt.xticks(df0,df1)
plt.yticks([100000,200000,300000,400000,500000,600000])
plt.xlabel('Month_number')
plt.ylabel('Sold units number')
plt.title('Graph 1')

```

```python
# ..................2 ...................#
plt.plot(df0.index,df1.values,linestyle='dotted',color='r',label='total profit',linewidth=3,marker='o')
plt.legend(loc='lower right') 
```

```python
#.................... 3 ..................#
product_sales=df.iloc[:,1:7]
product_sales
plt.plot(df0,product_sales,marker='s',mfc='g')
plt.xticks(df0)
plt.xlabel('Month number')
plt.ylabel('product sales')
plt.title('product sales-Graph')
plt.legend(loc='upper right')
plt.tight_layout()
```

```python
# .............. 4 ...............#
toothpaste_sales=df['toothpaste']
month_number=df['month_number']
plt.scatter(month_number,toothpaste_sales)
plt.xticks(month_number)
plt.xlabel('month_number')
plt.ylabel('toothpaste_sales')
```

```python
# ..............5 ...............#
face_cream=df['facecream']
face_wash=df['facewash']
plt.bar([a-0.25 for a in month_number],face_cream,width=0.25,align='edge',label='Face Cream Sales Data')
plt.bar([a+0.25 for a in month_number],face_wash,width=-0.25,align='edge',label='Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Per Month')
plt.title('Sales Data')
plt.xticks(month_number)
plt.legend(loc='upper right')
plt.tight_layout()
plt.grid()

```
