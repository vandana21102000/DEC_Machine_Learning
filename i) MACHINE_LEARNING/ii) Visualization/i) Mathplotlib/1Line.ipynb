```python
#array
import numpy as np
a=np.array([1,5,9,4,3])
b=np.array([5,7,2,1,7])
plt.plot(a,b)
```

```python
#if one axis is only defined
c=np.array([3,8,1,10,5,7]) #read as y coordinate
plt.plot(c)
#machine will read automatically the values as y coordinate and the machine assign the x values
```

```python
# * is act as a marker
plt.plot(c,'*')
```

```python
# o is act as a circle marker
plt.plot(c,'o')
```

```python
#line and marker 
plt.plot(c,marker='o')
```

```python
# marker,type of line,colour
y_points=np.array([1,5,2,7])
plt.plot(y_points,'*:g')
```

```python
#marker properties
x=np.array([1,5,2,7])
plt.plot(x,marker='o',ms=15,mfc='r',mec='g')
```

```python
#lines properties
plt.plot(c,linewidth=7,linestyle='dashed',color='c',marker='o',ms=20,mfc='k',mec='r')
```

```python
#Multiple lines in a graph
x1=[1,2,3,4]
x2=[5,6,7,8]
y1=[1,8,7,2]
y2=[5,7,2,3]
plt.plot(x1,y1,linewidth=7,linestyle='dashed',color='g',marker='h',ms=20,mfc='k',mec='r')
plt.plot(x2,y2,linewidth=7,linestyle='dashed',color='k',marker='o',ms=20,mfc='g',mec='c')
```

```python
#naming x and y axis labels
height=[150,155,158,160]
weight=[50,55,60,58]
plt.plot(height,weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height - Weight Graph")
```

```python
Displacement=[15,30,28,60]
Velocity=[50,55,60,88]
plt.plot(Displacement,Velocity)
plt.xlabel("Displacement",size=13,color='g')
plt.ylabel("Velocity",size=13,color='c')
plt.title("Velocity - Displacement Graph")
plt.grid()
```

```python
#subplot()
#we split the graph into two parts, the first part is only for graph 1 and the second part is for graph 2

x1=[1,2,3,4]
y1=[1,8,7,2]
plt.subplot(1,2,1) 
#3 rd argument in (1,2,1) represent the position
#the first and second argument in (1,2,1) represents the output display
plt.plot(x1,y1,marker='o')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Graph1')

x2=[5,6,7,8]
y2=[5,7,2,3]
plt.subplot(1,2,2)
plt.plot(x2,y2,marker='o')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Graph2')

#for getting a common main title for both graph 1 and graph 2
plt.suptitle('Graph1 - Graph2')
```
