```python
# Missing value
df['trestbps']=df['trestbps'].fillna(df['trestbps'].mean())
df['restecg']=df['restecg'].fillna(df['restecg'].mode()[0])
df['thalach']=df['thalach'].fillna(df['thalach'].mean())
df['ca']=df['ca'].fillna(df['ca'].mode()[0])
df['thal']=df['thal'].fillna(df['thal'].mode()[0])
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]  
x.values
```

```python
y=df.iloc[:,-1]
y.values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()   
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Model Creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)  
y_pred=knn.predict(x_test)  
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
label=[0,1]
cmd=ConfusionMatrixDisplay(cm,display_labels=label)
cmd.plot()
```

```python
#Hyper_parameter_tunning

from sklearn.model_selection import GridSearchCV
```

```python
#now we have to create a new object for Kneighbours classifier
model=KNeighborsClassifier()
```

```python
#assign the parameters used for tunning. We've to give the values as dictionary(key-value pair)

param={'n_neighbors': [3,5,7,9],'weights': ['uniform','distance']}

#Then we've to create object for GridSearchCV

gscv=GridSearchCV(model,param,cv=10,scoring='accuracy')
gscv.fit(x_train,y_train)  #(k=3,weights=uniform),(k=3,weights=distance),(k=5,weights=uniform),(k=5,weights=distance),........,(k=9,weights=uniform),(k=9,weights=distance)
```

```python
#To find the best parameter

print(gscv.best_params_)
```
