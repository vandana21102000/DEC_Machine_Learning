```python
df['Ram'] = df['Ram'].str.replace('GB','')
df['Weight'] = df['Weight'].str.replace('kg','')
df['Memory'] = df['Memory'].str.replace('GB','')
df['Memory'] = df['Memory'].str.replace('HDD','')
df['Memory'] = df['Memory'].str.replace('SSD','')
df['Memory'] = df['Memory'].str.replace('GB SSD','')
df['Memory'] = df['Memory'].str.replace('TB HDD','')
df['Memory'] = df['Memory'].str.replace('Hybrid','')
df['Memory'] = df['Memory'].str.replace('GB Hybrid','')
df['Memory'] = df['Memory'].str.replace('TB Hybrid','')
df['Memory'] = df['Memory'].str.replace('TB SSD','')
df['Memory'] = df['Memory'].str.replace('Flash Storage','')
df['Memory'] = df['Memory'].str.replace(r'[A-Za-z ]+', '', regex=True)
df['Memory'] = df['Memory'].str.replace('?','0')
df['Inches'] = df['Inches'].str.replace('?','0')
df
```
