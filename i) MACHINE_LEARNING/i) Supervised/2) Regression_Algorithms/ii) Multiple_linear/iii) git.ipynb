```python
#one hot encoding
#here we have to encode State
from sklearn.compose import make_column_transformer 
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
#make_column_transformer ==> transform the columns according to the number of labels
#handle_unknown='ignore'==> to avoid the unknown or the objects other than ('california','newyork','florida)
#remainder='passthrough'==>to pass all other columns without encoding except state column
x=col_trans.fit_transform(x)
x
```

```python
x.shape

#here now there will be 6 columns ('California','Florida','New York','R&D Spend', 'Administration', 'Marketing Spend')
#There will be no state column,it is divided into 3
#first the column is for 3 states then other columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
df5=pd.DataFrame({'actual_value':y_test,'predicted_value':y_pred,'Difference':y_test-y_pred})
df5
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,root_mean_squared_error,r2_score
print("mean absolute error",mean_absolute_error(y_test,y_pred))  
print("mean absolute percentage error",mean_absolute_percentage_error(y_test,y_pred))  #below .3 is good model
print("root mean squared error",root_mean_squared_error(y_test,y_pred))
print("r2_score",r2_score(y_test,y_pred))  #above 70% is good model
```
