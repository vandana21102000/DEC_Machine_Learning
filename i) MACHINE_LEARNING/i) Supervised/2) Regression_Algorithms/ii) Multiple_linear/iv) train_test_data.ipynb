```python
#Encoding
#get_dummies
#get dummies is an encoding method in pandas(not in machine learning)
dfe=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True,dtype=int)
dfe
```

```python
#Replacing unit(string)  ==> Remove strings from Mileage,Engine,Power
df_1['Mileage']=df_1['Mileage'].str.replace('km/kg','')
df_1['Mileage']=df_1['Mileage'].str.replace('kmpl','')
df_1['Engine']=df_1['Engine'].str.replace('CC','')
df_1['Power']=df_1['Power'].str.replace('bhp','')


#Mileage,Engine,power  ===> null to '0'
#in mileage,engine,power there is some null,so we've to replace it by 0.But we can't give it as a numerical value,because the dtype is still in object.
#  The 0 must be in string

#in mileage,Engine,Power columns there is 'null' typed in the data
#so we have to convert 'null to '0'('0' is entered as object not integer as the datatype of these columns are still object )

df_1['Mileage']=df_1['Mileage'].str.replace('null','0')
df_1['Engine']=df_1['Engine'].str.replace('null','0')
df_1['Power']=df_1['Power'].str.replace('null','0')
df_1
```

```python
df_1['Mileage']=df_1['Mileage'].str.replace('km/kg','')

df_1['Mileage']=df_1['Mileage'].str.replace('null','0')

df_1['Mileage']=df_1['Mileage'].astype(float)

df_1.loc[df_1.Mileage==0,'Mileage']=np.nan
```

```python
#dtype change
#we have to convert the object datatype integer into float

df_1['Mileage']=df_1['Mileage'].astype(float)
df_1['Engine']=df_1['Engine'].astype(float)
df_1['Power']=df_1['Power'].astype(float)
df_1
```

```python
#Mileage,Engine,Power ==>'null' ==>'0'
#now it is ==>numerical 0 (mileage cannot be 0 so we have to convert it to missing value(NaN))
df_1.loc[df_1.Engine==0,'Engine']=np.nan
df_1.loc[df_1.Mileage==0,'Mileage']=np.nan
df_1.loc[df_1.Power==0,'Power']=np.nan
```
