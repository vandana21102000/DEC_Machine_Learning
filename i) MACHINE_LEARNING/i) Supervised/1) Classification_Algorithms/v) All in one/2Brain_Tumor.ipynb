```python
#2. Encoding
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])
df['Tumor_Type']=lab.fit_transform(df['Tumor_Type'])
df['Location']=lab.fit_transform(df['Location'])
df['Histology']=lab.fit_transform(df['Histology'])
df['Stage']=lab.fit_transform(df['Stage'])
df['Symptom_1']=lab.fit_transform(df['Symptom_1'])
df['Symptom_2']=lab.fit_transform(df['Symptom_2'])
df['Symptom_3']=lab.fit_transform(df['Symptom_3'])
df['Radiation_Treatment']=lab.fit_transform(df['Radiation_Treatment'])
df['Surgery_Performed']=lab.fit_transform(df['Surgery_Performed'])
df['Chemotherapy']=lab.fit_transform(df['Chemotherapy'])
df['Family_History']=lab.fit_transform(df['Family_History'])
df['MRI_Result']=lab.fit_transform(df['MRI_Result'])
df
```

```python
df.drop(['Patient_ID'],axis=1,inplace=True)
df
```

```python
x=df.iloc[:,:-1]  
y=df.iloc[:,-1]
```

```python
#Converting data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
y_train.values
```

```python
# Normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()   #Object creation
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#3) MODEL CREATION
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
nav=GaussianNB()
sup=SVC()
lst=[knn,nav,sup]
```

```python
for i in lst:
    print("Model name is *******************************",i)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("Performance Evaluation is********************")
    print(confusion_matrix(y_test,y_pred))
    print('*'*100)
    print(accuracy_score(y_test,y_pred))
```
