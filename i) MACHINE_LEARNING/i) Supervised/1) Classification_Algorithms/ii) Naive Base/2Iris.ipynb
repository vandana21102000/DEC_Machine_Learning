```python
df.drop(['Id'],axis=1,inplace=True)
df
```

```python
x=df.iloc[:,:-1]  
y=df.iloc[:,-1]
```

```python
#Converting data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()   #Object creation
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Model creation
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
print(model.predict(scaler.transform([[4.8,3.2,5.5,2.5]])))
```

```python
#performance
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
label=['Iris-versicolor','Iris-virginica','Iris-setosa']
cmd=ConfusionMatrixDisplay(cm,display_labels=label)
cmd.plot()
```
