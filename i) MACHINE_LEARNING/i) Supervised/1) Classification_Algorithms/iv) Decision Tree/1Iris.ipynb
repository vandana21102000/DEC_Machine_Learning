```python
df.drop(['Id'],axis=1,inplace=True)
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
#Converting data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
#Model creation
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy') #have to add a parametrer (criterion='entropy')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
print(model.predict(([[4.8,3.2,5.5,2.5]])))
```

```python
#performance
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
label=['Iris-versicolor','Iris-virginica','Iris-setosa']
cmd=ConfusionMatrixDisplay(cm,display_labels=label)
cmd.plot()
```

```python
#plot Decision tree

from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
# tree.plot_tree(model)

#here the nodes are represented by x[0],x[1].. so we've to change it by the names.For this we use feature_names and for the leaf we use class_names (in the order of y train). For seeing each class in colour we use filled=True
tree.plot_tree(model,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=['Iris-versicolor','Iris-virginica','Iris-setosa'],filled=True)
```
